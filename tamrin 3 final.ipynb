```python
import os
documents = []
labels = []
for root, dirs, files in os.walk('/content/drive/MyDrive/bbcsport'):
    for filename in files:
        filepath = os.path.join(root, filename)
        file = open(filepath)
        content = file.read()
        documents.append(content)
        label = os.path.basename(root)
        labels.append(label)
```

```python
import pandas as pd
df=pd.DataFrame({'news':documents,'label':labels})
display(df)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()

x= vectorizer.fit_transform(df['news'])

```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, f1_score, balanced_accuracy_score
X=x
y=df['label']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
knn = KNeighborsClassifier(n_neighbors=3, metric='cosine')
knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)
res = accuracy_score(y_test, y_pred)
res1 = balanced_accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred, average='weighted')
print('Accuracy =', res)
print('Balanced Accuracy =', res1)
print('F1 Score =', f1)
```

```python
###########d=2
from sklearn.decomposition  import TruncatedSVD
svd = TruncatedSVD(n_components=2)
lsa = svd.fit_transform(x)
```
